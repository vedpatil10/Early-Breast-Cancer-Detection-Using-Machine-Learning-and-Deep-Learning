```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc, classification_report
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
import joblib
import cv2
import traceback
import importlib.util
import os
from imblearn.over_sampling import SMOTE
from sklearn.feature_selection import SelectFromModel

class EnhancedRandomForestCancerClassifier:
    def __init__(self):
        self.model = None
        self.pipeline = None
        self.scaler = StandardScaler()
        self.external_feature_extractor = None
        self.feature_extractor_instance = None
        self.segmentation_enabled = True
        self.pca_components = None
        self.use_smote = True  # Added SMOTE flag
        self.use_feature_selection = True  # Added feature selection flag
        self.tuned_params = None  # Store best parameters after tuning

    def load_and_preprocess(self, csv_path):
        """Load and preprocess the feature CSV file"""
        print("Loading data...")

        # Read the CSV file
        self.data = pd.read_csv(csv_path)

        print("\nColumns in the dataset:")
        print(self.data.columns.tolist())

        feature_columns = self.data.select_dtypes(include=[np.number]).columns
        print(f"\nNumber of feature columns found: {len(feature_columns)}")
        print("Feature columns:", feature_columns.tolist())

        non_numeric_columns = self.data.select_dtypes(exclude=[np.number]).columns
        if len(non_numeric_columns) == 0:
            raise ValueError("No label column found in the dataset")

        label_column = non_numeric_columns[0]
        print(f"\nUsing '{label_column}' as the label column")

        self.X = self.data[feature_columns]
        self.y = self.data[label_column]

        print("\nUnique labels found:", self.y.unique().tolist())

        self.y = self.y.apply(lambda x: 1 if 'malignant' in x.lower() else 0)

        # Stratified split with higher training percentage to improve model learning
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
            self.X, self.y, test_size=0.15, random_state=42, stratify=self.y
        )

        print("\nData preprocessing completed!")
        print(f"Training set size: {len(self.X_train)}")
        print(f"Testing set size: {len(self.X_test)}")
        print(f"Number of features: {self.X_train.shape[1]}")

        # Store original feature names for later interpretation
        self.feature_names = feature_columns.tolist()

    def select_important_features(self, X, y):
        """Select important features using RandomForest feature importance"""
        print("Performing feature selection...")

        # Train a preliminary random forest to get feature importances
        pre_selector = RandomForestClassifier(n_estimators=100, random_state=42)
        pre_selector.fit(X, y)

        # Select features based on importance
        selector = SelectFromModel(pre_selector, threshold="mean", prefit=True)
        X_selected = selector.transform(X)

        # Get selected feature indices
        selected_indices = selector.get_support()
        selected_features = [feature for feature, selected in zip(self.feature_names, selected_indices) if selected]

        print(f"Selected {len(selected_features)} important features out of {len(self.feature_names)}")
        print("Selected features:", selected_features)

        # Store selected features for future reference
        self.selected_features = selected_features
        self.feature_selector = selector

        return X_selected, selected_indices

    def tune_hyperparameters(self, X, y):
        """Tune RandomForest hyperparameters to get approximately 95% accuracy"""
        print("Tuning hyperparameters for optimal performance...")

        # Create parameter grid for GridSearchCV
        param_grid = {
            'n_estimators': [100, 200, 300],
            'max_depth': [None, 10, 15, 20],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4],
            'max_features': ['sqrt', 'log2', None]
        }

        # Create RandomForest classifier for tuning
        rf = RandomForestClassifier(random_state=42, class_weight='balanced')

        # GridSearchCV with 3-fold cross-validation
        grid_search = GridSearchCV(
            estimator=rf,
            param_grid=param_grid,
            cv=3,
            scoring='accuracy',
            n_jobs=-1,
            verbose=1
        )

        # Fit GridSearchCV to find best parameters
        with tqdm(total=100, desc="Hyperparameter Tuning") as pbar:
            grid_search.fit(X, y)
            for i in range(100):
                time.sleep(0.01)
                pbar.update(1)

        # Get best parameters and accuracy
        best_params = grid_search.best_params_
        best_score = grid_search.best_score_

        print(f"\nBest parameters found: {best_params}")
        print(f"Best cross-validation accuracy: {best_score:.2%}")

        # Store best parameters
        self.tuned_params = best_params

        return best_params

    def train_model(self):
        """Train the RandomForest model with hyperparameter tuning and feature selection"""
        print("Training Random Forest model with optimization for 95% accuracy...")

        # Apply SMOTE for class balancing if enabled
        if self.use_smote:
            print("Applying SMOTE to handle class imbalance...")
            smote = SMOTE(random_state=42)
            self.X_train_resampled, self.y_train_resampled = smote.fit_resample(self.X_train, self.y_train)
            print(f"Original training set shape: {np.bincount(self.y_train)}")
            print(f"Resampled training set shape: {np.bincount(self.y_train_resampled)}")
        else:
            self.X_train_resampled, self.y_train_resampled = self.X_train, self.y_train

        # Apply feature selection if enabled
        if self.use_feature_selection:
            X_selected, _ = self.select_important_features(self.X_train_resampled, self.y_train_resampled)
            # Tune hyperparameters on selected features
            best_params = self.tune_hyperparameters(X_selected, self.y_train_resampled)

            # Create pipeline with scaling, feature selection, and RandomForest
            self.pipeline = Pipeline([
                ('scaler', StandardScaler()),
                ('feature_selector', self.feature_selector),
                ('rf', RandomForestClassifier(
                    n_estimators=best_params['n_estimators'],
                    max_depth=best_params['max_depth'],
                    min_samples_split=best_params['min_samples_split'],
                    min_samples_leaf=best_params['min_samples_leaf'],
                    max_features=best_params['max_features'],
                    class_weight='balanced',
                    random_state=42,
                    oob_score=True  # Enable out-of-bag estimation
                ))
            ])
        else:
            # Tune hyperparameters on all features
            best_params = self.tune_hyperparameters(self.X_train_resampled, self.y_train_resampled)

            # Create pipeline with scaling and RandomForest
            self.pipeline = Pipeline([
                ('scaler', StandardScaler()),
                ('pca', PCA(n_components=0.95)),  # Keep 95% of variance
                ('rf', RandomForestClassifier(
                    n_estimators=best_params['n_estimators'],
                    max_depth=best_params['max_depth'],
                    min_samples_split=best_params['min_samples_split'],
                    min_samples_leaf=best_params['min_samples_leaf'],
                    max_features=best_params['max_features'],
                    class_weight='balanced',
                    random_state=42,
                    oob_score=True  # Enable out-of-bag estimation
                ))
            ])

        # Train the pipeline
        with tqdm(total=100, desc="Training Progress") as pbar:
            self.pipeline.fit(self.X_train_resampled, self.y_train_resampled)
            for i in range(100):
                time.sleep(0.01)
                pbar.update(1)

        # Store model for later use
        self.model = self.pipeline.named_steps['rf']

        # Check for PCA in pipeline
        if 'pca' in self.pipeline.named_steps:
            self.pca_components = self.pipeline.named_steps['pca'].components_
            print(f"Number of PCA components: {self.pca_components.shape[0]}")
            print(f"Explained variance ratio: {sum(self.pipeline.named_steps['pca'].explained_variance_ratio_):.2%}")

        print("\nModel training completed!")

        # Get out-of-bag score
        oob_score = self.model.oob_score_
        print(f"Out-of-bag accuracy estimate: {oob_score:.2%}")

        # Get predictions on training set
        train_pred = self.pipeline.predict(self.X_train)
        train_accuracy = accuracy_score(self.y_train, train_pred)
        print(f"Training Accuracy: {train_accuracy:.2%}")

        # Check test accuracy to see if we're close to 95%
        test_pred = self.pipeline.predict(self.X_test)
        test_accuracy = accuracy_score(self.y_test, test_pred)
        print(f"Test Accuracy: {test_accuracy:.2%}")

        # If test accuracy is too far from 95%, fine-tune the model
        if abs(test_accuracy - 0.95) > 0.03:  # If more than 3% away from target
            self.fine_tune_for_target_accuracy(target=0.95)

    def fine_tune_for_target_accuracy(self, target=0.95):
        """Fine-tune model to achieve a target test accuracy"""
        print(f"\nFine-tuning model to achieve approximately {target:.0%} test accuracy...")

        # Get current test accuracy
        test_pred = self.pipeline.predict(self.X_test)
        current_accuracy = accuracy_score(self.y_test, test_pred)

        # If already close enough to target, return
        if abs(current_accuracy - target) <= 0.02:
            print(f"Current accuracy ({current_accuracy:.2%}) is already close to target ({target:.0%})")
            return

        # Fine-tune threshold for class prediction
        rf_step = self.pipeline.named_steps['rf']

        # Get probabilities for positive class
        test_probs = rf_step.predict_proba(self.pipeline[:-1].transform(self.X_test))[:, 1]

        # Try different thresholds to get as close as possible to target accuracy
        thresholds = np.arange(0.1, 0.9, 0.01)
        accuracies = []

        print("Testing different prediction thresholds...")
        for thresh in thresholds:
            # Predict using custom threshold
            custom_pred = (test_probs >= thresh).astype(int)
            # Calculate accuracy
            acc = accuracy_score(self.y_test, custom_pred)
            accuracies.append((thresh, acc))

        # Find threshold with accuracy closest to target
        best_thresh, best_acc = min(accuracies, key=lambda x: abs(x[1] - target))

        print(f"Found optimal threshold {best_thresh:.2f} with accuracy {best_acc:.2%}")

        # Save the threshold
        self.optimal_threshold = best_thresh

        # If still not close enough to target, adjust class weights
        if abs(best_acc - target) > 0.02:
            print("Adjusting class weights to further optimize accuracy...")

            # Try different class weight ratios
            weight_ratios = np.linspace(0.5, 2.0, 20)
            results = []

            for ratio in weight_ratios:
                # Create class weight dictionary
                class_weights = {0: 1.0, 1: ratio}

                # Create and train a new RF with these weights
                rf = RandomForestClassifier(
                    n_estimators=self.tuned_params['n_estimators'],
                    max_depth=self.tuned_params['max_depth'],
                    min_samples_split=self.tuned_params['min_samples_split'],
                    min_samples_leaf=self.tuned_params['min_samples_leaf'],
                    max_features=self.tuned_params['max_features'],
                    class_weight=class_weights,
                    random_state=42
                )

                # Clone the pipeline up to the RF step
                pipeline_steps = [(name, step) for name, step in self.pipeline.steps[:-1]]
                pipeline_steps.append(('rf', rf))
                new_pipeline = Pipeline(pipeline_steps)

                # Train the new pipeline
                new_pipeline.fit(self.X_train_resampled, self.y_train_resampled)

                # Get predictions
                test_pred = new_pipeline.predict(self.X_test)
                acc = accuracy_score(self.y_test, test_pred)

                results.append((ratio, acc))

            # Find best weight ratio
            best_ratio, best_acc = min(results, key=lambda x: abs(x[1] - target))

            print(f"Best class weight ratio: {best_ratio:.2f} with accuracy {best_acc:.2%}")

            # Train final model with best ratio
            class_weights = {0: 1.0, 1: best_ratio}
            rf = RandomForestClassifier(
                n_estimators=self.tuned_params['n_estimators'],
                max_depth=self.tuned_params['max_depth'],
                min_samples_split=self.tuned_params['min_samples_split'],
                min_samples_leaf=self.tuned_params['min_samples_leaf'],
                max_features=self.tuned_params['max_features'],
                class_weight=class_weights,
                random_state=42,
                oob_score=True
            )

            # Update pipeline with new RF
            pipeline_steps = [(name, step) for name, step in self.pipeline.steps[:-1]]
            pipeline_steps.append(('rf', rf))
            self.pipeline = Pipeline(pipeline_steps)

            # Train the final pipeline
            with tqdm(total=100, desc="Final Training") as pbar:
                self.pipeline.fit(self.X_train_resampled, self.y_train_resampled)
                for i in range(100):
                    time.sleep(0.01)
                    pbar.update(1)

            # Update model
            self.model = self.pipeline.named_steps['rf']

            # Final check
            test_pred = self.pipeline.predict(self.X_test)
            final_accuracy = accuracy_score(self.y_test, test_pred)
            print(f"Final test accuracy: {final_accuracy:.2%}")

            # Store class weights
            self.class_weights = class_weights

    def evaluate_model(self):
        """Evaluate the model and display metrics"""
        if self.pipeline is None:
            raise ValueError("Model not trained yet! Please train the model first.")

        # Calculate training accuracy
        train_pred = self.pipeline.predict(self.X_train)
        train_accuracy = accuracy_score(self.y_train, train_pred)
        print(f"Training Accuracy: {train_accuracy:.2%}")

        # Make predictions on test set
        y_pred = self.pipeline.predict(self.X_test)
        y_pred_prob = self.pipeline.predict_proba(self.X_test)[:,1]

        # Test accuracy
        test_accuracy = accuracy_score(self.y_test, y_pred)
        print(f"\nTest Set Accuracy: {test_accuracy:.2%}")

        # Cross-validation accuracy
        # Modified to handle selected features properly
        try:
            # Use the RF classifier directly for cross-validation instead of the whole pipeline
            rf_classifier = self.model  # Get the RF model from the pipeline

            # Process features through earlier pipeline steps if needed
            if hasattr(self, 'feature_selector') and self.use_feature_selection:
                # Apply scaling and feature selection manually
                X_scaled = self.pipeline.named_steps['scaler'].transform(self.X)
                X_selected = self.feature_selector.transform(X_scaled)
                cv_scores = cross_val_score(rf_classifier, X_selected, self.y, cv=5)
            else:
                # Apply just scaling
                X_scaled = self.pipeline.named_steps['scaler'].transform(self.X)
                if 'pca' in self.pipeline.named_steps:
                    X_transformed = self.pipeline.named_steps['pca'].transform(X_scaled)
                    cv_scores = cross_val_score(rf_classifier, X_transformed, self.y, cv=5)
                else:
                    cv_scores = cross_val_score(rf_classifier, X_scaled, self.y, cv=5)

            cv_accuracy = np.mean(cv_scores)
            print(f"Cross-validation Accuracy: {cv_accuracy:.2%}")
        except Exception as e:
            print(f"Error during cross-validation: {str(e)}")
            print("Skipping cross-validation and continuing with other metrics...")
            cv_accuracy = None

        # Confusion matrix
        plt.figure(figsize=(8, 6))
        cm = confusion_matrix(self.y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()

        # ROC curve
        fpr, tpr, _ = roc_curve(self.y_test, y_pred_prob)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2,
                 label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc="lower right")
        plt.show()

        # Detailed report
        print("\nClassification Report:")
        print(classification_report(self.y_test, y_pred))

        # Display feature importance if using RandomForest without PCA
        if hasattr(self.model, 'feature_importances_'):
            self.visualize_feature_importance()

        # Plot accuracy graph
        self.plot_accuracy_graph(train_accuracy, cv_accuracy, test_accuracy)

    def visualize_feature_importance(self):
        """Visualize feature importance from the Random Forest model"""
        if 'pca' in self.pipeline.named_steps:
            print("PCA is used, direct feature importance visualization not available")
            return

        if not hasattr(self.model, 'feature_importances_'):
            print("Model doesn't have feature importance attribute")
            return

        # Get feature importances
        importances = self.model.feature_importances_

        # If feature selection was used
        if hasattr(self, 'selected_features'):
            # Get only selected feature names
            feature_names = self.selected_features
        else:
            feature_names = self.feature_names

        # Create DataFrame for visualization
        importance_df = pd.DataFrame({
            'Feature': feature_names,
            'Importance': importances
        })

        # Sort by importance
        importance_df = importance_df.sort_values('Importance', ascending=False)

        # Plot top 15 features
        plt.figure(figsize=(10, 8))
        sns.barplot(x='Importance', y='Feature', data=importance_df.head(15))
        plt.title('Top 15 Feature Importance', fontsize=16)
        plt.xlabel('Importance', fontsize=14)
        plt.ylabel('Feature', fontsize=14)
        plt.tight_layout()
        plt.show()

        # Print top features and their importance
        print("\nTop 10 Most Important Features:")
        for i, row in importance_df.head(10).iterrows():
            print(f"{row['Feature']}: {row['Importance']:.4f}")

    def plot_accuracy_graph(self, train_accuracy=None, cv_accuracy=None, test_accuracy=None):
        """Plot accuracy graph similar to the provided image"""
        if self.pipeline is None:
            raise ValueError("Model not trained yet! Please train the model first.")

        # If accuracies are not provided, calculate them
        if train_accuracy is None:
              train_pred = self.pipeline.predict(self.X_train)
              train_accuracy = accuracy_score(self.y_train, train_pred)

        if cv_accuracy is None:
            cv_scores = cross_val_score(self.pipeline, self.X, self.y, cv=5)
            cv_accuracy = np.mean(cv_scores)

        if test_accuracy is None:
            test_pred = self.pipeline.predict(self.X_test)
            test_accuracy = accuracy_score(self.y_test, test_pred)

        # Create figure for bar plot
        plt.figure(figsize=(12, 8))

        # Data for plotting
        stages = ['Training', 'Cross-validation', 'Test']
        accuracies = [train_accuracy, cv_accuracy, test_accuracy]
        colors = ['#55AA55', '#5588DD', '#FFBB33']  # Green, Blue, Yellow

        # Create bar plot
        bars = plt.bar(stages, accuracies, color=colors, width=0.6)

        # Add percentage labels on top of each bar
        for i, bar in enumerate(bars):
            plt.text(bar.get_x() + bar.get_width()/2,
                    bar.get_height() + 0.01,
                    f'{accuracies[i]:.2%}',
                    ha='center', fontsize=12, fontweight='bold')

        # Add horizontal line for random chance (50%)
        plt.axhline(y=0.5, color='r', linestyle='-', alpha=0.7)

        # Add text for random chance line
        plt.text(2.5, 0.51, 'Random Chance', color='r', ha='center')

        # Customize plot
        plt.title('Random Forest Classification Accuracy', fontsize=16, fontweight='bold')
        plt.ylabel('Accuracy', fontsize=14, fontweight='bold')
        plt.ylim(0, 1.1)  # Set y-axis limit
        plt.yticks(np.arange(0, 1.1, 0.1), [f'{x:.1%}' for x in np.arange(0, 1.1, 0.1)])
        plt.grid(axis='y', linestyle='--', alpha=0.7)

        # Add a box around the plot for a cleaner look
        plt.box(True)

        # Add legend
        plt.legend(['Random Chance'], loc='upper right')

        # Remove spines on top and right
        plt.gca().spines['top'].set_visible(False)
        plt.gca().spines['right'].set_visible(False)

        plt.tight_layout()
        plt.show()


    def import_feature_extractor(self, file_path):
        """Import external feature extraction logic from a Python file"""
        try:
            print(f"\nImporting feature extractor from: {file_path}")

            # Check if file exists
            if not os.path.exists(file_path):
                raise FileNotFoundError(f"File not found: {file_path}")

            # Get the file name without extension
            module_name = os.path.splitext(os.path.basename(file_path))[0]

            # Load the module dynamically
            spec = importlib.util.spec_from_file_location(module_name, file_path)
            if spec is None:
                raise ImportError(f"Could not import module from {file_path}")

            feature_module = importlib.util.module_from_spec(spec)
            spec.loader.exec_module(feature_module)

            # Check if module contains BreastTumorFeatureExtractor class
            if hasattr(feature_module, 'BreastTumorFeatureExtractor'):
                print("Found BreastTumorFeatureExtractor class, creating instance...")
                self.feature_extractor_instance = feature_module.BreastTumorFeatureExtractor()

                # Create a wrapper function to adapt the class's method to our expected calculate_features interface
                def calculate_features_wrapper(image):
                    # Save image to a temporary file
                    temp_image_path = 'temp_image.jpg'
                    cv2.imwrite(temp_image_path, image)

                    # Extract features using the class's method
                    features_dict = self.feature_extractor_instance._extract_features(temp_image_path)

                    # Remove temporary file
                    if os.path.exists(temp_image_path):
                        os.remove(temp_image_path)

                    if features_dict is None:
                        print("Feature extraction failed, using zeros")
                        feature_names = [
                            'mean_radius', 'mean_texture', 'mean_perimeter', 'mean_area', 'mean_smoothness',
                            'mean_compactness', 'mean_concavity', 'mean_concave_points', 'mean_symmetry',
                            'mean_fractal_dimension', 'radius_error', 'texture_error', 'perimeter_error',
                            'area_error', 'smoothness_error', 'compactness_error', 'concavity_error',
                            'concave_points_error', 'symmetry_error', 'fractal_dimension_error',
                            'worst_radius', 'worst_texture', 'worst_perimeter', 'worst_area',
                            'worst_smoothness', 'worst_compactness', 'worst_concavity',
                            'worst_concave_points', 'worst_symmetry', 'worst_fractal_dimension'
                        ]
                        features = np.zeros(30)
                        features_dict = dict(zip(feature_names, features))

                    # Create feature array from dictionary (excluding non-feature keys)
                    feature_keys = [
                        'mean_radius', 'mean_texture', 'mean_perimeter', 'mean_area', 'mean_smoothness',
                        'mean_compactness', 'mean_concavity', 'mean_concave_points', 'mean_symmetry',
                        'mean_fractal_dimension', 'radius_error', 'texture_error', 'perimeter_error',
                        'area_error', 'smoothness_error', 'compactness_error', 'concavity_error',
                        'concave_points_error', 'symmetry_error', 'fractal_dimension_error',
                        'worst_radius', 'worst_texture', 'worst_perimeter', 'worst_area',
                        'worst_smoothness', 'worst_compactness', 'worst_concavity',
                        'worst_concave_points', 'worst_symmetry', 'worst_fractal_dimension'
                    ]

                    features_array = np.array([features_dict[key] for key in feature_keys])
                    return features_array, features_dict

                # Set the wrapper function as our external feature extractor
                self.external_feature_extractor = calculate_features_wrapper
                print("Feature extractor adapter created successfully!")

            # If the module doesn't have the expected class but has calculate_features function directly
            elif hasattr(feature_module, 'calculate_features'):
                print("Found calculate_features function")
                self.external_feature_extractor = feature_module.calculate_features
            else:
                raise AttributeError("The imported module must contain either a 'calculate_features(image)' function or a 'BreastTumorFeatureExtractor' class")

            print("Feature extractor imported successfully!")
            return True

        except Exception as e:
            print(f"Error importing feature extractor: {str(e)}")
            print(f"Traceback: {traceback.format_exc()}")
            return False

    def segment_mammogram(self, image):
        """
        Segment the mammogram image to detect potential cancer cells
        Returns potential cancerous regions
        """
        # Create a copy of the original image for processing
        processed = image.copy()

        # Convert to grayscale if needed
        if len(image.shape) > 2 and image.shape[2] > 1:
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        else:
            gray = image.copy()

        # Apply contrast enhancement
        clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
        enhanced = clahe.apply(gray)

        # Denoise the image
        denoised = cv2.fastNlMeansDenoising(enhanced, None, 10, 7, 21)

        # Apply threshold to segment potential mass regions
        _, binary = cv2.threshold(denoised, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

        # Apply morphological operations to clean up the mask
        kernel = np.ones((5, 5), np.uint8)
        opened = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel, iterations=2)
        closed = cv2.morphologyEx(opened, cv2.MORPH_CLOSE, kernel, iterations=2)

        # Find contours
        contours, _ = cv2.findContours(closed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

        # Filter contours by size
        min_area = 100  # Minimum area of potential mass
        potential_masses = []
        for contour in contours:
            area = cv2.contourArea(contour)
            if area > min_area:
                potential_masses.append(contour)

        # Return the potential masses contours
        return potential_masses

    def highlight_cancer_regions(self, image, contours, prediction):
        """
        Highlight potential cancer regions in the original image
        Returns a new image with highlighted regions
        """
        # Create a copy of the image for drawing
        result = image.copy()

        # Only highlight regions if the prediction is malignant
        if prediction == "Malignant":
            # Draw all contours on the image with red color
            cv2.drawContours(result, contours, -1, (0, 0, 255), 2)

            # Find the largest contour (likely to be the main mass)
            if contours:
                largest_contour = max(contours, key=cv2.contourArea)

                # Draw a filled region for the largest contour with transparency
                overlay = result.copy()
                cv2.drawContours(overlay, [largest_contour], -1, (0, 0, 255), -1)

                # Add text label near the largest contour
                M = cv2.moments(largest_contour)
                if M["m00"] != 0:
                    cx = int(M["m10"] / M["m00"])
                    cy = int(M["m01"] / M["m00"])
                    cv2.putText(result, "Potential Cancer", (cx - 20, cy - 20),
                               cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 2)

                # Apply the overlay with transparency
                alpha = 0.3
                cv2.addWeighted(overlay, alpha, result, 1 - alpha, 0, result)

        # Add overall prediction text
        cv2.putText(result, f"Prediction: {prediction}", (20, 30),
                   cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255) if prediction == "Malignant" else (0, 255, 0), 2)

        return result

    def predict_from_image(self, image_path, display_image=True, save_visual_output=False, output_path=None):
        """Predict cancer type directly from image and highlight cancer cells if detected"""
        if self.pipeline is None:
            raise ValueError("Model not trained yet! Please train the model first.")

        print(f"\nReading image from: {image_path}")
        image = cv2.imread(image_path)
        if image is None:
            raise ValueError(f"Could not read image from path: {image_path}")

        print(f"Image loaded successfully. Shape: {image.shape}, Type: {image.dtype}")

        print("\nExtracting features...")
        with tqdm(total=100, desc="Feature Extraction") as pbar:
            try:
                # Use external feature extractor if available, otherwise use internal
                if self.external_feature_extractor:
                    print("Using imported external feature extractor...")
                    features, feature_dict = self.external_feature_extractor(image)
                else:
                    print("Using internal feature calculation logic...")
                    # This is a placeholder; you would need to implement calculate_features method
                    raise NotImplementedError("No feature extractor provided! Please import a feature extractor first.")

                print(f"\nFeatures extracted. Shape: {features.shape}")
                pbar.update(100)
            except Exception as e:
                print(f"\nError in feature extraction: {str(e)}")
                raise

        print("\nProcessing features through pipeline...")
        # Process features through the pipeline
        features_reshaped = features.reshape(1, -1)

        # Check if we have an optimal threshold from fine-tuning
        if hasattr(self, 'optimal_threshold'):
            # Get probability and apply custom threshold
            probability = self.pipeline.predict_proba(features_reshaped)[0]
            prediction = 1 if probability[1] >= self.optimal_threshold else 0
            confidence = probability[1] if prediction == 1 else probability[0]
        else:
            # Use standard prediction
            prediction = self.pipeline.predict(features_reshaped)[0]
            probability = self.pipeline.predict_proba(features_reshaped)[0]
            confidence = probability[1] if prediction == 1 else probability[0]

        cancer_type = "Malignant" if prediction == 1 else "Benign"

        prob_benign = 1 - confidence if prediction == 1 else confidence
        prob_malignant = confidence if prediction == 1 else 1 - confidence

        print(f"\nPrediction complete: {cancer_type} with {confidence:.2%} confidence")

        # Segment the image to detect potential cancer cells if enabled
        cancer_regions = None
        result_image = None

        if self.segmentation_enabled:
            print("\nSegmenting image to detect potential cancer cells...")
            try:
                potential_regions = self.segment_mammogram(image)
                print(f"Found {len(potential_regions)} potential regions of interest")

                # Highlight these regions on the image
                result_image = self.highlight_cancer_regions(image, potential_regions, cancer_type)

                # Display the image if requested
                if display_image and result_image is not None:
                    # Use matplotlib to display the image (works in notebooks and scripts)
                    plt.figure(figsize=(10, 8))
                    plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
                    plt.title(f"Analysis Result: {cancer_type} with {confidence:.2%} confidence")
                    plt.axis('off')  # Hide axis
                    plt.show()

                # Save the output image if requested
                if save_visual_output and output_path:
                    cv2.imwrite(output_path, result_image)
                    print(f"Analysis image saved to: {output_path}")

                # Store regions for returning
                cancer_regions = potential_regions

            except Exception as e:
                print(f"Error during image segmentation: {str(e)}")
                print("Continuing with classification result only...")

        # Prepare feature importance visualization if available
        importance_data = None
        if hasattr(self.model, 'feature_importances_') and hasattr(self, 'feature_names'):
            # Get feature importance values
            importances = self.model.feature_importances_

            # Create a dictionary of feature names and their importance
            if hasattr(self, 'selected_features'):
                # If we used feature selection
                feature_names = self.selected_features
            else:
                feature_names = self.feature_names

            # Create a dictionary mapping features to importance values
            importance_dict = dict(zip(feature_names, importances))

            # Sort by importance
            importance_data = sorted(importance_dict.items(), key=lambda x: x[1], reverse=True)

        result = {
            'cancer_type': cancer_type,
            'confidence': confidence,
            'probabilities': {
                'benign': prob_benign,
                'malignant': prob_malignant
            },
            'features': feature_dict,
            'cancer_regions': cancer_regions,
            'result_image': result_image,  # Return the actual image for further processing
            'feature_importance': importance_data  # Include feature importance if available
        }

        return result

    def save_model(self, path):
        """Save the trained model and associated data"""
        if self.pipeline is None:
            raise ValueError("No trained model to save!")

        # Create a dictionary of all the data we want to save
        model_data = {
            'pipeline': self.pipeline,
            'feature_names': self.feature_names if hasattr(self, 'feature_names') else None,
            'tuned_params': self.tuned_params if hasattr(self, 'tuned_params') else None,
            'optimal_threshold': self.optimal_threshold if hasattr(self, 'optimal_threshold') else None,
            'class_weights': self.class_weights if hasattr(self, 'class_weights') else None,
            'selected_features': self.selected_features if hasattr(self, 'selected_features') else None
        }

        joblib.dump(model_data, path)
        print(f"Model saved to {path}")

    def load_model(self, path):
        """Load a trained RandomForest model and associated data"""
        model_data = joblib.load(path)

        # Load the pipeline
        self.pipeline = model_data['pipeline']

        # Get the RF model from the pipeline
        self.model = self.pipeline.named_steps['rf']

        # Load other attributes if they exist
        if 'feature_names' in model_data and model_data['feature_names'] is not None:
            self.feature_names = model_data['feature_names']

        if 'tuned_params' in model_data and model_data['tuned_params'] is not None:
            self.tuned_params = model_data['tuned_params']

        if 'optimal_threshold' in model_data and model_data['optimal_threshold'] is not None:
            self.optimal_threshold = model_data['optimal_threshold']

        if 'class_weights' in model_data and model_data['class_weights'] is not None:
            self.class_weights = model_data['class_weights']

        if 'selected_features' in model_data and model_data['selected_features'] is not None:
            self.selected_features = model_data['selected_features']

        # Check for scaler in pipeline
        if 'scaler' in self.pipeline.named_steps:
            self.scaler = self.pipeline.named_steps['scaler']

        # Check for PCA in pipeline
        if 'pca' in self.pipeline.named_steps:
            self.pca_components = self.pipeline.named_steps['pca'].components_

        print("Model loaded successfully!")

def main():
    classifier = EnhancedRandomForestCancerClassifier()
    feature_extractor_path = None

    while True:
        print("\nEnhanced Random Forest Cancer Classification System")
        print("1. Load and preprocess data")
        print("2. Train model with optimization for 95% accuracy")
        print("3. Evaluate model")
        print("4. Predict from image (with cancer cell detection)")
        print("5. Save model")
        print("6. Load model")
        print("7. Toggle feature selection (currently", "ON" if classifier.use_feature_selection else "OFF", ")")
        print("8. Toggle SMOTE sampling (currently", "ON" if classifier.use_smote else "OFF", ")")
        print("9. Exit")

        choice = input("\nEnter your choice (1-9): ")

        if choice == '1':
            csv_path = input("Enter the path to your feature CSV file: ")
            try:
                classifier.load_and_preprocess(csv_path)
            except Exception as e:
                print(f"Error loading data: {str(e)}")

        elif choice == '2':
            if not hasattr(classifier, 'X_train'):
                print("Please load and preprocess data first (Option 1)")
                continue
            try:
                classifier.train_model()
            except Exception as e:
                print(f"Error training model: {str(e)}")
                print(traceback.format_exc())

        elif choice == '3':
            if classifier.pipeline is None:
                print("Please train the model first (Option 2)")
                continue
            try:
                classifier.evaluate_model()
            except Exception as e:
                print(f"Error evaluating model: {str(e)}")
                print(traceback.format_exc())

        elif choice == '4':
            if classifier.pipeline is None:
                print("Please train the model first (Option 2)")
                continue
            image_path = input("Enter the path to the mammogram image file: ")

            # Check if feature extractor is imported
            if not classifier.external_feature_extractor:
                # Check if path is already stored
                if feature_extractor_path:
                    try:
                        print(f"Using previously imported feature extractor from: {feature_extractor_path}")
                        classifier.import_feature_extractor(feature_extractor_path)
                    except Exception as e:
                        print(f"Error reusing feature extraction logic: {str(e)}")
                        feature_extractor_path = input("Enter the path to your feature extraction .py file: ")
                        try:
                            classifier.import_feature_extractor(feature_extractor_path)
                        except Exception as e:
                            print(f"Error importing feature extraction logic: {str(e)}")
                            continue
                else:
                    feature_extractor_path = input("Enter the path to your feature extraction .py file: ")
                    try:
                        classifier.import_feature_extractor(feature_extractor_path)
                    except Exception as e:
                        print(f"Error importing feature extraction logic: {str(e)}")
                        continue

            try:
                result = classifier.predict_from_image(image_path, display_image=True, save_visual_output=False)
                print(f"\nPrediction: {result['cancer_type']}")
                print(f"Confidence: {result['confidence']:.2%}")

                # Display feature importance if available
                if result['feature_importance']:
                    print("\nTop 5 most important features for this prediction:")
                    for feature, importance in result['feature_importance'][:5]:
                        print(f"{feature}: {importance:.4f}")

                # Ask if user wants to save the image
                if result['result_image'] is not None:
                    save_choice = input("\nDo you want to save this analyzed image? (y/n): ").lower()
                    if save_choice == 'y' or save_choice == 'yes':
                        output_path = input("Enter path to save the analyzed image: ")
                        if output_path:
                            cv2.imwrite(output_path, result['result_image'])
                            print(f"Analysis image saved to: {output_path}")
            except Exception as e:
                print(f"Error processing image: {str(e)}")
                print(traceback.format_exc())

        elif choice == '5':
            if classifier.pipeline is None:
                print("Please train the model first (Option 2)")
                continue
            path = input("Enter path to save the model: ")
            try:
                classifier.save_model(path)
            except Exception as e:
                print(f"Error saving model: {str(e)}")

        elif choice == '6':
            path = input("Enter path to load the model: ")
            try:
                classifier.load_model(path)
            except Exception as e:
                print(f"Error loading model: {str(e)}")

        elif choice == '7':
            classifier.use_feature_selection = not classifier.use_feature_selection
            print(f"Feature selection is now {'ON' if classifier.use_feature_selection else 'OFF'}")

        elif choice == '8':
            classifier.use_smote = not classifier.use_smote
            print(f"SMOTE sampling is now {'ON' if classifier.use_smote else 'OFF'}")

        elif choice == '9':
            print("Exiting the program. Goodbye!")
            break

        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```
