```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import time
import joblib
import cv2
import traceback
import tensorflow as tf
from tensorflow.keras.models import Sequential, load_model, Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Input, GlobalAveragePooling2D
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from imblearn.over_sampling import SMOTE
from tensorflow.keras import mixed_precision

policy = mixed_precision.Policy('mixed_float16')
mixed_precision.set_global_policy(policy)


class CancerClassifierCNN:
    def __init__(self, input_shape=(128, 128, 3)):  # Changed from 224x224 to 128x128
        self.model = None
        self.input_shape = input_shape
        self.history = None

    def build_model(self, use_pretrained=True):
        """Build a CNN model, either from scratch or using pretrained MobileNetV2"""
        if use_pretrained:
            # Use MobileNetV2 as base model
            base_model = MobileNetV2(
                input_shape=self.input_shape,
                include_top=False,
                weights='imagenet'
            )

            # Freeze the base model layers
            base_model.trainable = False

            # Add custom classification layers
            inputs = Input(shape=self.input_shape)
            x = tf.keras.applications.mobilenet_v2.preprocess_input(inputs)
            x = base_model(x, training=False)
            x = GlobalAveragePooling2D()(x)
            x = BatchNormalization()(x)
            x = Dropout(0.5)(x)
            x = Dense(128, activation='relu')(x)
            x = Dropout(0.3)(x)
            outputs = Dense(1, activation='sigmoid')(x)

            self.model = Model(inputs=inputs, outputs=outputs)
            print("Created model based on MobileNetV2")

        else:
            # Build a custom CNN model from scratch
            self.model = Sequential([
                # First convolutional block
                Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=self.input_shape),
                BatchNormalization(),
                Conv2D(32, (3, 3), activation='relu', padding='same'),
                BatchNormalization(),
                MaxPooling2D((2, 2)),
                Dropout(0.25),

                # Second convolutional block
                Conv2D(64, (3, 3), activation='relu', padding='same'),
                BatchNormalization(),
                Conv2D(64, (3, 3), activation='relu', padding='same'),
                BatchNormalization(),
                MaxPooling2D((2, 2)),
                Dropout(0.25),

                # Third convolutional block
                Conv2D(128, (3, 3), activation='relu', padding='same'),
                BatchNormalization(),
                Conv2D(128, (3, 3), activation='relu', padding='same'),
                BatchNormalization(),
                MaxPooling2D((2, 2)),
                Dropout(0.25),

                # Classifier
                Flatten(),
                Dense(512, activation='relu'),
                BatchNormalization(),
                Dropout(0.5),
                Dense(1, activation='sigmoid')  # Binary classification
            ])
            print("Created custom CNN model from scratch")

        # Compile the model
        self.model.compile(
            optimizer=Adam(learning_rate=0.0001),
            loss='binary_crossentropy',
            metrics=['accuracy', tf.keras.metrics.AUC()]
        )

        # Print model summary
        self.model.summary()
        return self.model

    def load_images_from_directory(self, directory, label_mapping=None):
        """
        Load images from directory structure.
        Assumes directory contains subdirectories for each class.
        """
        print("Loading images from directory...")

        images = []
        labels = []
        file_paths = []

        # If no label mapping provided, infer from directory names
        if label_mapping is None:
            label_mapping = {d: i for i, d in enumerate(os.listdir(directory))
                           if os.path.isdir(os.path.join(directory, d))}

        print(f"Found classes: {label_mapping}")

        for class_name, class_label in label_mapping.items():
            class_dir = os.path.join(directory, class_name)
            if not os.path.isdir(class_dir):
                print(f"Warning: {class_dir} is not a directory, skipping")
                continue

            print(f"Processing class: {class_name}, label: {class_label}")

            # Process all images in the class directory
            image_files = [f for f in os.listdir(class_dir)
                          if f.lower().endswith(('.png', '.jpg', '.jpeg'))]

            for img_file in tqdm(image_files, desc=f"Loading {class_name}"):
                try:
                    img_path = os.path.join(class_dir, img_file)
                    image = cv2.imread(img_path)
                    if image is None:
                        print(f"Warning: Could not read {img_path}, skipping")
                        continue

                    # Resize image to expected input shape (now 128x128)
                    image = cv2.resize(image, (self.input_shape[0], self.input_shape[1]))
                    # Convert BGR to RGB
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                    images.append(image)
                    labels.append(class_label)
                    file_paths.append(img_path)

                except Exception as e:
                    print(f"Error processing {img_file}: {str(e)}")
                    continue

        # Convert to numpy arrays
        X = np.array(images)
        y = np.array(labels)

        print(f"Loaded {len(X)} images with shape {self.input_shape}")
        print(f"Label distribution: {np.bincount(y)}")

        return X, y, file_paths

    def load_and_preprocess(self, data_path, binary_classification=True, apply_smote=False):
        """Load and preprocess images from a directory or CSV file"""
        if os.path.isdir(data_path):
            # Load from directory structure where folder names are class names
            self.X, self.y, self.file_paths = self.load_images_from_directory(data_path)

            # For cancer classification, we might want to convert labels to binary
            if binary_classification and len(np.unique(self.y)) > 2:
                print("Converting to binary classification...")
                # Typically label 0 could be benign and others malignant, but you might need to adjust
                self.y = np.where(self.y == 0, 0, 1)
                print(f"New label distribution: {np.bincount(self.y)}")

        elif data_path.endswith('.csv'):
            # Load from CSV file that contains image paths and labels
            print("Loading from CSV file...")
            data = pd.read_csv(data_path)

            # Check if the CSV has the expected columns
            if 'image_path' not in data.columns or 'label' not in data.columns:
                raise ValueError("CSV file must contain 'image_path' and 'label' columns")

            images = []
            labels = []
            file_paths = []

            for _, row in tqdm(data.iterrows(), total=len(data), desc="Loading images"):
                try:
                    img_path = row['image_path']
                    if not os.path.exists(img_path):
                        print(f"Warning: {img_path} does not exist, skipping")
                        continue

                    image = cv2.imread(img_path)
                    if image is None:
                        print(f"Warning: Could not read {img_path}, skipping")
                        continue

                    # Resize image to expected input shape (now 128x128)
                    image = cv2.resize(image, (self.input_shape[0], self.input_shape[1]))
                    # Convert BGR to RGB
                    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

                    images.append(image)

                    # Convert label to binary if needed
                    label = row['label']
                    if binary_classification:
                        # Convert string labels to binary
                        if isinstance(label, str):
                            label = 1 if 'malignant' in label.lower() else 0

                    labels.append(label)
                    file_paths.append(img_path)

                except Exception as e:
                    print(f"Error processing row {_}: {str(e)}")
                    continue

            # Convert to numpy arrays
            self.X = np.array(images)
            self.y = np.array(labels)
            self.file_paths = file_paths

            print(f"Loaded {len(self.X)} images with shape {self.input_shape}")
            if binary_classification:
                print(f"Label distribution: {np.bincount(self.y)}")
            else:
                unique_labels, counts = np.unique(self.y, return_counts=True)
                print("Label distribution:", dict(zip(unique_labels, counts)))

        else:
            raise ValueError(f"Unsupported data path format: {data_path}. Use directory path or CSV file.")

        # Split data into training and testing sets
        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
            self.X, self.y, test_size=0.2, random_state=42, stratify=self.y
        )

        # Apply SMOTE to the training data if requested
        if apply_smote:
            print("\nApplying SMOTE to balance the training data...")
            # Reshape the images to 2D for SMOTE
            original_shape = self.X_train.shape
            X_reshaped = self.X_train.reshape(original_shape[0], -1)

            # Apply SMOTE
            sm = SMOTE(random_state=42)
            X_resampled, y_resampled = sm.fit_resample(X_reshaped, self.y_train)

            # Reshape back to original image dimensions
            self.X_train = X_resampled.reshape(-1, original_shape[1], original_shape[2], original_shape[3])
            self.y_train = y_resampled

            print(f"After SMOTE - Training set size: {len(self.X_train)}")
            print(f"New label distribution in training set: {np.bincount(self.y_train)}")

        print("\nData preprocessing completed!")
        print(f"Training set size: {len(self.X_train)}")
        print(f"Testing set size: {len(self.X_test)}")
        print(f"Input shape: {self.input_shape}")

    def train_model(self, epochs=20, batch_size=32, augment_data=True, use_class_weights=True):
        """Train the CNN model with data augmentation, callbacks, and class weights"""
        if self.model is None:
            raise ValueError("Model not built! Call build_model() first.")

        if not hasattr(self, 'X_train') or self.X_train is None:
            raise ValueError("Data not loaded! Call load_and_preprocess() first.")

        # Setup data augmentation
        if augment_data:
            datagen = ImageDataGenerator(
                rotation_range=20,
                width_shift_range=0.1,
                height_shift_range=0.1,
                shear_range=0.1,
                zoom_range=0.1,
                horizontal_flip=True,
                fill_mode='nearest',
                rescale=1./255  # Normalize pixel values
            )

            # Validation data only needs rescaling, not augmentation
            val_datagen = ImageDataGenerator(
                rescale=1./255
            )
        else:
            # Without augmentation, only apply rescaling
            datagen = ImageDataGenerator(rescale=1./255)
            val_datagen = ImageDataGenerator(rescale=1./255)

        # Configure callbacks
        callbacks = [
            EarlyStopping(
                monitor='val_loss',
                patience=10,
                restore_best_weights=True,
                verbose=1
            ),
            ReduceLROnPlateau(
                monitor='val_loss',
                factor=0.5,
                patience=5,
                min_lr=1e-6,
                verbose=1
            ),
            ModelCheckpoint(
                'best_cancer_model.h5',
                monitor='val_auc',
                mode='max',
                save_best_only=True,
                verbose=1
            )
        ]

        # Create generators
        train_generator = datagen.flow(
            self.X_train, self.y_train,
            batch_size=batch_size
        )

        validation_generator = val_datagen.flow(
            self.X_test, self.y_test,
            batch_size=batch_size
        )

        # Calculate steps per epoch
        steps_per_epoch = len(self.X_train) // batch_size
        validation_steps = len(self.X_test) // batch_size

        # Ensure at least 1 step
        steps_per_epoch = max(1, steps_per_epoch)
        validation_steps = max(1, validation_steps)

        # Calculate and apply class weights if requested
        class_weight_dict = None
        if use_class_weights:
            # Import necessary function
            from sklearn.utils.class_weight import compute_class_weight

            # Get class distribution
            unique_classes = np.unique(self.y_train)
            class_counts = np.bincount(self.y_train)
            print(f"\nClass distribution in training data: {class_counts}")

            # Calculate class weights - inverse of frequency
            class_weights = compute_class_weight(
                'balanced',
                classes=unique_classes,
                y=self.y_train
            )

            # Create dictionary mapping class indices to weights
            class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
            print(f"Using class weights: {class_weight_dict}")

        print("\nTraining CNN model with data augmentation" +
              (" and class weighting" if use_class_weights else "") + "...")

        # Train the model
        self.history = self.model.fit(
            train_generator,
            steps_per_epoch=steps_per_epoch,
            epochs=epochs,
            validation_data=validation_generator,
            validation_steps=validation_steps,
            callbacks=callbacks,
            class_weight=class_weight_dict,  # Apply class weights if available
            verbose=1
        )

        # Load the best weights
        self.model.load_weights('best_cancer_model.h5')

        print("\nModel training completed!")

        # Print final metrics
        train_loss, train_acc, train_auc = self.model.evaluate(
            datagen.flow(self.X_train, self.y_train, batch_size=batch_size),
            steps=steps_per_epoch,
            verbose=0
        )

        val_loss, val_acc, val_auc = self.model.evaluate(
            val_datagen.flow(self.X_test, self.y_test, batch_size=batch_size),
            steps=validation_steps,
            verbose=0
        )

        print(f"Training Accuracy: {train_acc:.2%}, AUC: {train_auc:.4f}")
        print(f"Validation Accuracy: {val_acc:.2%}, AUC: {val_auc:.4f}")

        return self.history

    def evaluate_model(self, batch_size=16):
        """Evaluate the model and display metrics with memory efficiency"""
        if self.model is None:
            raise ValueError("Model not trained yet! Please train the model first.")

        if not hasattr(self, 'X_test') or self.X_test is None:
            raise ValueError("Test data not available! Load data first.")

        print("\nEvaluating model in batches to conserve memory...")

        # Create a generator for prediction to avoid memory issues
        datagen = ImageDataGenerator(rescale=1./255)
        test_generator = datagen.flow(
            self.X_test, self.y_test,
            batch_size=batch_size, shuffle=False
        )

        # Get predictions in batches
        y_pred_prob = []
        steps = len(self.X_test) // batch_size + (1 if len(self.X_test) % batch_size > 0 else 0)

        for i in range(steps):
            batch_x, _ = next(test_generator)
            batch_pred = self.model.predict(batch_x, verbose=0)
            y_pred_prob.extend(batch_pred)

        y_pred_prob = np.array(y_pred_prob).flatten()
        y_pred = (y_pred_prob > 0.5).astype(int)

        # Calculate accuracy
        accuracy = accuracy_score(self.y_test, y_pred)
        print(f"\nTest Set Accuracy: {accuracy:.2%}")

        # Plot confusion matrix
        plt.figure(figsize=(8, 6))
        cm = confusion_matrix(self.y_test, y_pred)
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()

        # Plot ROC curve
        fpr, tpr, _ = roc_curve(self.y_test, y_pred_prob)
        roc_auc = auc(fpr, tpr)

        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='darkorange', lw=2,
                label=f'ROC curve (AUC = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc="lower right")
        plt.show()

        # Print detailed metrics
        print("\nDetailed Performance Metrics:")
        print(f"True Negatives (Correct Benign): {cm[0,0]}")
        print(f"False Positives (Incorrect Malignant): {cm[0,1]}")
        print(f"False Negatives (Incorrect Benign): {cm[1,0]}")
        print(f"True Positives (Correct Malignant): {cm[1,1]}")

        # Clear memory
        import gc
        gc.collect()

        # Plot training history if available
        if hasattr(self, 'history') and self.history is not None:
            plt.figure(figsize=(12, 5))

            # Plot training & validation accuracy
            plt.subplot(1, 2, 1)
            plt.plot(self.history.history['accuracy'])
            plt.plot(self.history.history['val_accuracy'])
            plt.title('Model Accuracy')
            plt.ylabel('Accuracy')
            plt.xlabel('Epoch')
            plt.legend(['Train', 'Validation'], loc='upper left')

            # Plot training & validation loss
            plt.subplot(1, 2, 2)
            plt.plot(self.history.history['loss'])
            plt.plot(self.history.history['val_loss'])
            plt.title('Model Loss')
            plt.ylabel('Loss')
            plt.xlabel('Epoch')
            plt.legend(['Train', 'Validation'], loc='upper left')

            plt.tight_layout()
            plt.show()

    def predict_from_image(self, image_path):
        """Predict cancer type directly from image"""
        if self.model is None:
            raise ValueError("Model not trained yet! Please train the model first.")

        # Read the image
        print(f"\nReading image from: {image_path}")
        image = cv2.imread(image_path)
        if image is None:
            raise ValueError(f"Could not read image from path: {image_path}")

        print(f"Image loaded successfully. Shape: {image.shape}, Type: {image.dtype}")

        # Preprocess the image
        image = cv2.resize(image, (self.input_shape[0], self.input_shape[1]))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Normalize
        image = image / 255.0

        # Add batch dimension
        image = np.expand_dims(image, axis=0)

        # Make prediction
        print("Making prediction...")
        probability = self.model.predict(image)[0][0]
        prediction = 1 if probability > 0.5 else 0

        # Format result
        cancer_type = "Malignant" if prediction == 1 else "Benign"
        confidence = probability if prediction == 1 else 1 - probability

        print(f"\nPrediction complete: {cancer_type} with {confidence:.2%} confidence")

        # Visualize the image and prediction
        plt.figure(figsize=(8, 6))
        plt.imshow(image[0])
        plt.title(f"Prediction: {cancer_type} ({confidence:.2%} confidence)")
        plt.axis('off')
        plt.show()

        # Return detailed results
        return {
            'cancer_type': cancer_type,
            'confidence': float(confidence),
            'probabilities': {
                'benign': float(1 - probability),
                'malignant': float(probability)
            }
        }

    def explain_prediction(self, image_path):
        """Explain the prediction using Grad-CAM visualization"""
        if self.model is None:
            raise ValueError("Model not trained yet! Please train the model first.")

        try:
            import tensorflow.keras.backend as K
            from tensorflow.keras.models import Model

            # Read and preprocess the image
            image = cv2.imread(image_path)
            image = cv2.resize(image, (self.input_shape[0], self.input_shape[1]))
            image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
            input_image = image / 255.0

            # Make prediction on the original image
            input_batch = np.expand_dims(input_image, axis=0)
            prediction = self.model.predict(input_batch)[0][0]
            predicted_class = 1 if prediction > 0.5 else 0

            # Find the last convolutional layer
            last_conv_layer = None
            for layer in reversed(self.model.layers):
                if isinstance(layer, tf.keras.layers.Conv2D):
                    last_conv_layer = layer.name
                    break

            if last_conv_layer is None:
                raise ValueError("Could not find a convolutional layer in the model")

            # Create a model that outputs both the predictions and the activations of the last conv layer
            grad_model = Model(
                inputs=self.model.inputs,
                outputs=[self.model.get_layer(last_conv_layer).output, self.model.output]
            )

            # Get the gradients of the predicted class with respect to the output feature map
            with tf.GradientTape() as tape:
                conv_output, predictions = grad_model(input_batch)
                loss = predictions[:, 0]  # For binary classification

            # Extract gradients
            grads = tape.gradient(loss, conv_output)

            # Global average pooling on the gradients
            pooled_grads = K.mean(grads, axis=(0, 1, 2))

            # Compute the heatmap by weighing each channel and summing
            heatmap = tf.reduce_mean(tf.multiply(pooled_grads, conv_output), axis=-1)
            heatmap = tf.squeeze(heatmap)
            heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
            heatmap = heatmap.numpy()

            # Resize heatmap to match original image size
            heatmap = cv2.resize(heatmap, (self.input_shape[0], self.input_shape[1]))

            # Convert heatmap to RGB
            heatmap = np.uint8(255 * heatmap)
            heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

            # Superimpose the heatmap on original image
            superimposed_img = heatmap * 0.4 + image
            superimposed_img = np.clip(superimposed_img, 0, 255).astype('uint8')

            # Display the original image and the Grad-CAM visualization
            plt.figure(figsize=(12, 5))

            plt.subplot(1, 2, 1)
            plt.imshow(image)
            plt.title(f"Original: {'Malignant' if predicted_class == 1 else 'Benign'} ({prediction:.2%})")
            plt.axis('off')

            plt.subplot(1, 2, 2)
            plt.imshow(superimposed_img)
            plt.title("Grad-CAM: Areas of Interest")
            plt.axis('off')

            plt.tight_layout()
            plt.show()

            return {
                'prediction': 'Malignant' if predicted_class == 1 else 'Benign',
                'confidence': float(prediction if predicted_class == 1 else 1 - prediction),
                'heatmap': superimposed_img
            }

        except Exception as e:
            print(f"Error generating Grad-CAM visualization: {str(e)}")
            traceback.print_exc()
            return None

    def save_model(self, path):
        """Save the trained model"""
        if self.model is None:
            raise ValueError("No trained model to save!")

        # Save model
        self.model.save(path)
        print(f"Model saved to {path}")

    def load_model(self, path):
        """Load a trained model"""
        self.model = load_model(path)
        print("Model loaded successfully!")

def main():
    classifier = CancerClassifierCNN()  # Now initializes with 128x128 input shape

    while True:
        print("\nCancer Classification CNN System")
        print("1. Build model")
        print("2. Load and preprocess data")
        print("3. Train model")
        print("4. Evaluate model")
        print("5. Predict from image")
        print("6. Explain prediction")
        print("7. Save model")
        print("8. Load model")
        print("9. Exit")

        choice = input("\nEnter your choice (1-9): ")

        if choice == '1':
            use_pretrained = input("Use pretrained MobileNetV2? (y/n): ").lower() == 'y'
            try:
                classifier.build_model(use_pretrained=use_pretrained)
            except Exception as e:
                print(f"Error building model: {str(e)}")
                traceback.print_exc()

        elif choice == '2':
            data_path = input("Enter path to data directory or CSV file: ")
            apply_smote = input("Apply SMOTE to balance classes? (y/n): ").lower() == 'y'
            try:
                classifier.load_and_preprocess(data_path, apply_smote=apply_smote)
            except Exception as e:
                print(f"Error loading data: {str(e)}")
                traceback.print_exc()
                print("\nPlease make sure your data:")
                print("1. Is in a directory with class subdirectories OR")
                print("2. Is a CSV file with 'image_path' and 'label' columns")

        elif choice == '3':
            if classifier.model is None:
                print("Please build the model first (Option 1)")
                continue
            if not hasattr(classifier, 'X_train'):
                print("Please load and preprocess data first (Option 2)")
                continue

            try:
                epochs = int(input("Enter number of epochs (default: 20): ") or "20")
                batch_size = int(input("Enter batch size (default: 32): ") or "32")
                augment = input("Use data augmentation? (y/n, default: y): ").lower() != 'n'

                classifier.train_model(epochs=epochs, batch_size=batch_size, augment_data=augment)
            except Exception as e:
                print(f"Error training model: {str(e)}")
                traceback.print_exc()

        elif choice == '4':
            if classifier.model is None:
                print("Please train the model first (Option 3)")
                continue
            try:
                classifier.evaluate_model()
            except Exception as e:
                print(f"Error evaluating model: {str(e)}")
                traceback.print_exc()

        elif choice == '5':
            if classifier.model is None:
                print("Please train the model first (Option 3)")
                continue
            image_path = input("Enter the path to the image file: ")
            try:
                result = classifier.predict_from_image(image_path)
                print(f"\nPrediction: {result['cancer_type']}")
                print(f"Confidence: {result['confidence']:.2%}")
                print("\nProbabilities:")
                print(f"- Benign: {result['probabilities']['benign']:.2%}")
                print(f"- Malignant: {result['probabilities']['malignant']:.2%}")
            except Exception as e:
                print(f"Error processing image: {str(e)}")
                traceback.print_exc()

        elif choice == '6':
            if classifier.model is None:
                print("Please train the model first (Option 3)")
                continue
            image_path = input("Enter the path to the image file: ")
            try:
                classifier.explain_prediction(image_path)
            except Exception as e:
                print(f"Error explaining prediction: {str(e)}")
                traceback.print_exc()

        elif choice == '7':
            if classifier.model is None:
                print("Please train the model first (Option 3)")
                continue
            path = input("Enter path to save the model: ")
            try:
                classifier.save_model(path)
            except Exception as e:
                print(f"Error saving model: {str(e)}")
                traceback.print_exc()

        elif choice == '8':
            path = input("Enter path to load the model: ")
            try:
                classifier.load_model(path)
            except Exception as e:
                print(f"Error loading model: {str(e)}")
                traceback.print_exc()

        elif choice == '9':
            print("Done!")
            break

        else:
            print("Invalid choice! Please try again.")

if __name__ == "__main__":
    main()
```
